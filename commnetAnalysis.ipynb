```python
import nltk
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download the VADER lexicon and other required resources
nltk.download('vader_lexicon')

# Load your dataset (replace 'your_dataset.csv' with your actual file)
data = pd.read_csv('/content/UScomments.csv', error_bad_lines=False)

# Create an instance of the SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

# Perform sentiment analysis on each comment in the dataset
sentiments = []

for comment in data['comment_text']:
    if isinstance(comment, float):  # Check if the comment is a float
        comment = str(comment)       # Convert float to string
    sentiment_scores = sid.polarity_scores(comment)

    if sentiment_scores['compound'] >= 0.05:
        sentiment = 'positive'
    elif sentiment_scores['compound'] <= -0.05:
        sentiment = 'negative'
    else:
        sentiment = 'neutral'

    sentiments.append(sentiment)

```

```python


# Add the 'sentiment' column to the dataset
data['sentiment'] = sentiments

# Print a sample of the dataset with sentiments
print(data[['comment_text', 'sentiment']].head())

# Test the sentiment analysis on sample comments
sample_comments = ["I love this product!", "This is terrible.", "It's okay, I guess."]
sample_sentiments = []

for comment in sample_comments:
    sentiment_scores = sid.polarity_scores(comment)

    if sentiment_scores['compound'] >= 0.05:
        sentiment = 'positive'
    elif sentiment_scores['compound'] <= -0.05:
        sentiment = 'negative'
    else:
        sentiment = 'neutral'

    sample_sentiments.append(sentiment)

# Print the sentiments of the sample comments
print("\nSample Comment Sentiments:")
for i, comment in enumerate(sample_comments):
    print(f"Comment: '{comment}' | Sentiment: {sample_sentiments[i]}")

```
